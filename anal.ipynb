```python
MP=pd.read_excel('Downloads/프로야구 연봉.xlsx',parse_cols='A, B, C, E, F',encoding='urf-8')
```

```python
MP.head()
```

```python
plt.figure(figsize=(30,30))
plt.scatter(MP['국내 선수 연봉(만원)'], MP['WAR'], s=100)
plt.xlabel('국내 선수 연봉(만원)')
plt.ylabel('WAR')
plt.grid()
plt.show()
```

```python
import numpy as np
```

```python
fp1=np.polyfit(MP['국내 선수 연봉(만원)'], MP['WAR'], 1)
fp1
```

```python
f1=np.poly1d(fp1)
fx=np.linspace(8000, 300000, 10)
```

```python
plt.figure(figsize=(30,30))
plt.scatter(MP['국내 선수 연봉(만원)'], MP['WAR'], s=100)
plt.plot(fx,f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('국내 선수 연봉(만원)')
plt.ylabel('WAR')
plt.grid()
plt.show()
```

```python
fp1=np.polyfit(MP['국내 선수 연봉(만원)'], MP['WAR'], 1)

f1=np.poly1d(fp1)
fx=np.linspace(8000, 300000, 10)

MP['오차'] = np.abs(MP['WAR']-f1(MP['국내 선수 연봉(만원)']))

df_sort = MP.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(30,30))
plt.scatter(MP['국내 선수 연봉(만원)'], MP['WAR'], c=MP['오차'], s=50)
plt.plot(fx,f1(fx), ls='dashed', lw=3, color='g')

for n in range(100):
    plt.text(df_sort['국내 선수 연봉(만원)'][n]*1.02, df_sort['WAR'][n]*0.98, df_sort['이름'][n], fontsize=10)
    
plt.xlabel('국내 선수 연봉(만원)')
plt.ylabel('WAR')

plt.colorbar()
plt.grid()
plt.show()
```
